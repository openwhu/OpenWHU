建立一个波包，它不会像我们在图6.2中观察到的高斯波包那样分裂成两部分，而是均匀地朝一个方向移动。提示：到目前为止，在本节讨论的模拟中，我们假设弦在$t=0$之前是静止的。为了构造一个不会立即分裂的单一波包，你必须正确地指定弦的初始位移和速度。

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from IPython.display import HTML, display, clear_output

# 参数设置
L = 1.0    # 弦的长度
v = 0.2    # 波包的速度
m = 1.0    # 波包的质量
hbar = 1.0 # 约化普朗克常数
sigma = 0.05    # 波包展宽

# 初始化位置坐标
x = np.linspace(0, L, 1000)

# 计算初始波矢和初始位移
k0 = m * v / hbar
x0 = L / 4

# 构造波包函数
def gaussian_wave_packet(x, x0, k0, sigma, t):
    psi = np.exp(-(x - x0 - v * t)**2 / (2 * sigma**2)) * np.exp(1j * k0 * (x - v * t))
    return psi

# 动画更新函数
def update(frame):
    t = frame / 50.0     # 时间步长
    psi = gaussian_wave_packet(x, x0, k0, sigma, t)
    line.set_ydata(np.abs(psi)**2)
    return line,

# 创建画布和图像对象
fig, ax = plt.subplots(figsize=(8,6))
line, = ax.plot(x, np.abs(gaussian_wave_packet(x, x0, k0, sigma, 0))**2)
ax.set_xlabel('Position')
ax.set_ylabel('Probability density')
ax.set_title('Time Evolution of Gaussian Wave Packet')

# 创建动画
ani = animation.FuncAnimation(fig, update, frames=200, interval=50, blit=False)


# 显示动画
display(HTML(ani.to_html5_video()))
plt.close(fig)
```
